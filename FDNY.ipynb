```python
import pandas as pd

# Read the CSV file "FDNY.csv" from the "Downloads" directory, skipping the first row
df = pd.read_csv("Downloads/14.FDNY.csv", skiprows=1)

# Display the DataFrame
print(df)
```

```python
# Display basic statistics for numerical columns
print(df.describe())
```

```python
print(df.columns)
```

```python
print(df.index)
```

```python
for column in df.columns:
    count = df[column].count()
    print(f"Column '{column}' has {count} records")
```

```python
import pandas as pd

# Assuming 'df' is your DataFrame

# List of columns you want to select and display
selected_columns = ['FacilityName', 'FacilityAddress', 'Borough']

# Group the data by 'Borough' and select the specified columns for each borough
borough_wise_info = df.groupby('Borough')[selected_columns].apply(lambda x: x.reset_index(drop=True))

# Display the borough-wise information
print(borough_wise_info)
```

```python
import pandas as pd

# Assuming 'df' is your DataFrame

# Get the unique boroughs in the DataFrame
unique_boroughs = df['Borough'].unique()

# Loop through each unique borough and display information
for borough in unique_boroughs:
    borough_info = df[df['Borough'] == borough]
    print(f"Information for Borough: {borough}")
    print(borough_info)
    print('\n')  # Add a newline for separation
```
